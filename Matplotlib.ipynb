```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4]
dsa=[30,40,50,60]
ml=[70,80,90,100]

plt.plot(x,dsa, label="DSA Marks",color='purple',linewidth="2",marker='*',linestyle="--",markersize='10')
plt.plot(x,ml,'r*--' ,label="ML Marks")  #r*-- is positional shorthand property it will placed in between only.

plt.title("Marks Insights" ,
          fontdict={'fontsize':'20','fontname':'Calibri','fontstyle':'italic','color':'blue'})
plt.xlabel("No.of Hours Studying", fontdict={'fontsize':'10','fontname':'Arial','fontstyle':'oblique','color':'red'})
plt.ylabel("Marks", fontdict={'fontsize':'10','fontname':'Arial','fontstyle':'oblique','color':'red'})

plt.legend(loc="lower right")
plt.show()
           
    
```
